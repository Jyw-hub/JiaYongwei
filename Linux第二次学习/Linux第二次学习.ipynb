#### which-find指令

which 指令名字

find 初始查找路径 --name “被查找文件名路径”

find 初始查找路径 -siza ＋-n[KMG]

+表示大于，-表示小于，n表示单位，K表示KB，M表示MB，G表示GB，B表示B


#### echo-tail-重定向符

echo 输出内容  echo  `  `

例如 echo ` pwd ` 会把当前目录输出

echo 内容 > 文件名  重定向符

例如 echo “hello world” > hello.txt 会把hello world覆盖写入hello.txt

echo 内容 >> 文件名  

例如 echo “hello world” >> hello.txt 会把hello world追加写入hello.txt

ls -l > 文件名

例如 ls -l > hello.txt 会把当前目录的ls -l输出覆盖写入hello.txt

tail [-f num] Linux路径名

-f表示持续追踪

num表示显示的倒数层数，默认倒数十层

#### grep-wc-管道符

$ grep [-n] 关键字 文件路径(可以作为管道的输入)

-n表示显示匹配行号

例如：

grep -n “hello” hello.txt 

会显示hello.txt中包含hello的行数

$ wc [-c -m -l -w ] 文件路径

-c表示字节数         

-m表示字符数

-l表示行数

-w表示单词数

例如：

wc -l hello.txt 

会显示hello.txt的行数 

不带选项时候，会显示行数，单词数，字节数，文件名

管道符 |  可以嵌套多个命令

$ 命令1 | 命令2

命令1的输出作为命令2的输入


#### vim-vi-模式

i 进入编辑模式

a 进入追加模式            

ESC 退出编辑模式

:wq 保存并退出

:q! 强制退出不保存

:set nu 显示行号


```python
from matplotlib import font_manager

my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc', size=18)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


def show_image(image_path):
    try:
        img = mpimg.imread(image_path)
        plt.imshow(img)
        plt.axis('off')  # 隐藏坐标轴
        plt.title("显示图片", fontproperties=my_font)
        plt.show()
    except FileNotFoundError:
        print(f"错误：文件 {image_path} 不存在！")
    except Exception as e:
        print(f"加载失败: {e}")


# 使用示例
show_image("sample6.png")
show_image("sample1.png")
show_image("sample2.png")
show_image("sample3.png")
show_image("sample4.png")
show_image("sample5.png")
```

#### 用户和用户组管理

useradd [-g -d] 用户名  # 创建用户

使用-g选项指定用户组，如果不指定-g则默认创建名同名的组。-d选项指定用户主目录，如果不指定，则默认创建/home/用户名目录

userdel [-r] 用户名  # 删除用户


-r选项表示删除用户目录,不使用-r选项则仅删除用户

passwd 用户名    # 修改用户密码

groupadd 组名  # 创建用户组

groupdel 组名  # 删除用户组

usermod -aG 组名mod 用户名   # 将用户加入用户组

id 用户名
