#### $ ls [-a -l -h] [目录名]


-a 显示所有文件，包括隐藏文件。

-l 显示详细信息，包括文件权限、所有者、大小、修改日期等。

-h 显示文件大小用更易读的格式。要搭配-l选项一起使用。

如果不指定目录名，则默认为当前目录。


!(./sample1.png)  # 假设图片在 Notebook 所在目录的 images 文件夹中

```python
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc', size=18)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_image(image_path):
    try:
        img = mpimg.imread(image_path)
        plt.imshow(img)
        plt.axis('off')  # 隐藏坐标轴
        plt.title("显示图片", fontproperties=my_font)
        plt.show()
    except FileNotFoundError:
        print(f"错误：文件 {image_path} 不存在！")
    except Exception as e:
        print(f"加载失败: {e}")

# 使用示例
show_image("sample1.png")
show_image("sample2.png")
```
